```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt 
```

```python
dataset=pd.read_csv('cancer_reg.csv')
```

```python
dataset = dataset.drop('pctsomecol18_24', axis=1) # Remove pctsomecol18_24 
dataset = dataset.drop('binnedinc', axis=1) # Remove pctsomecol18_24 
dataset = dataset.drop('geography', axis=1) # Remove pctsomecol18_24 
```

```python
df = dataset.astype('float32')
print(df.dtypes)
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler() # scale the data using minmax scaler
scaler.fit(df)
df[:] = scaler.fit_transform(df[:])
df_scaled = pd.DataFrame(df[:], columns=df.columns[:])
df_scaled
```

```python
X = df_scaled.drop(['target_deathrate'], axis=1)
y = df_scaled['target_deathrate']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = MLPRegressor(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred_nn_train = model.predict(X_train)
y_pred_nn_test = model.predict(X_test)

r2_nn_train = r2_score(y_train, y_pred_nn_train)
print("Training R^2 for NN Model: ", r2_nn_train*100)

r2_nn_test = r2_score(y_test, y_pred_nn_test)
print("Testing R^2 for NN Model: ", r2_nn_test*100)

RMSE_nn_train = sqrt(mean_squared_error(y_train, y_pred_nn_train))
print("RMSE for Training Data: ", RMSE_nn_train)

RMSE_nn_test = sqrt(mean_squared_error(y_test, y_pred_nn_test))
print("RMSE for Testing Data: ", RMSE_nn_test)
```

```python
true_value = y_test
predicted_value = y_pred_nn_test
```

```python
plt.figure(figsize=(8,8))
plt.scatter(true_value, predicted_value, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(predicted_value), max(true_value))
p2 = min(min(predicted_value), min(true_value))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predicted', fontsize=15)
plt.title("NN : True Values vs Predicted ")
plt.axis('equal')
plt.show()
```
